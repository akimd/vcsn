# _context_.trie(data="", format="default", filename="")

Generate a "trie" automaton (a prefix tree) from a finite series, given as a file of (weighted) words.

Arguments:
- _data_: a string containing the list of words
- _format_:
  - "default": same as "monomials"
  - "monomials": each line contains a weighted word in the monomials syntax: `<2>foo`
  - "words": each line contains a single word: `foo`.  In this case `<2>foo` is read as a five-letter word.

Postconditions:
- `Result.is_deterministic()`

See also:
- [context.cotrie](context.cotrie.ipynb)
- [polynomial.cotrie](polynomial.cotrie.ipynb)
- [polynomial.trie](polynomial.trie.ipynb)

## Examples

### Words (finite language)

```python
import vcsn
vcsn.B.trie('''foo
bar
baz''')
```

```python
%%file words
hello
world
hell
word
```

```python
vcsn.B.trie(filename='words')
```

### Weighted words (finite series)

```python
vcsn.Q.trie('''
one
<2>two
<3>three
<13>thirteen
<30>thirty
<51>thirsty''')
```

### Tuples of words

```python
vcsn.context('lat<law, law>, q').trie('''
<1>one|un
<2>two|deux
<3>three|trois
<4>four|quatre
<14>forteen|quatorze
<40>forty|quarante''')
```
