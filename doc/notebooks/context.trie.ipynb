# _context_.trie(_`file`_)

Generate a "trie" automaton (a prefix tree) from a finite series, given as a file of (weighted) words.

Postconditions:
- `Result.is_deterministic()`

See also:
- [polynomial.trie](polynomial.trie.ipynb)

## Examples

```python
import vcsn
```

### Words (finite language)

```python
%%file words
hello
world
hell
word
```

```python
vcsn.B.trie('words')
```

### Weighted words (finite series)

```python
%%file numbers
one
<2>two
<3>three
<4>four
```

```python
vcsn.Z.trie('numbers')
```

### Tuples of words

```python
%%file translator
one,un
two,deux
three,trois
four, quatre
forteen, quatorze
forty, quarante
```

```python
t = vcsn.context('lat<law_char, law_char>, b').trie('translator')
t
```
