Turn an automaton into a spontaneous automaton by mapping each transition $\lmul{w}{\ell}$ into $\lmul{\lmul{w}{\ell}}{\varepsilon}$.

Preconditions:
- None

See also:
- [expression.lift](expression.lift.ipynb)

## Examples

```python
import vcsn
a = vcsn.Z.expression('(<1>a+<2>b+<3>c){2}').automaton()
a
```

```python
b = a.lift()
b
```

```python
from IPython.display import display
display(a.context(), b.context())
```

A different set of identities can be chosen.

```python
a.lift().type()
```

```python
a.lift(identities = 'trivial').type()
```
