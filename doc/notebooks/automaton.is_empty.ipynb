# _automaton_.is_empty

Whether the automaton is _empty_, i.e., it has no states.

See also:
- [_automaton_.is_useless](automaton.is_useless.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton a dot
digraph
{
  vcsn_context = "lal_char, b"
  0
}
```

This automaton is useless (it accepts the empty language), yet it is not empty.

```python
a.is_empty()
```

If we trim it, then it is empty (which is why it seems that there is no result):

```python
a.trim()
```

```python
a.trim().is_empty()
```
