# _automaton_.is_proper

Whether the automaton is proper, i.e. has no spontaneous transitions.

See also:

## Examples

```python
import vcsn
exp = vcsn.context('lan, b').expression('ab')
```

```python
a = exp.standard()
a
```

```python
a.is_proper()
```
