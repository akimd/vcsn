# _automaton_.lightest_automaton(algo = "auto")

Return an automaton containing only the transitions and states from the lightest path in `automaton`.

The algorithm can be: 
- `"auto"`: uses `"dijkstra"` implementation if the weightset cannot have lightening weights (for example $\mathbb{N}_{\text{min}}$), "bellman-ford" otherwise.
- `"a-star"`
- `"bellman-ford"`
- `"dijkstra"`

Preconditions:
- `automaton` must be tropical.

See also:
- [_automaton_.lightest](automaton.lightest.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton --strip aut
context = "lal_char(abcd), nmin"
$ -> 0
0 -> 1 <6>a
0 -> 2 <1>a
2 -> 3 <1>b
3 -> 3 <2>b
3 -> 4 <1>c
4 -> 1 <1>d
0 -> 5 <2>a
5 -> 1 <3>b
1 -> $
```

```python
aut.lightest_automaton()
```

```python
aut.lightest_automaton("bellman-ford")
```
