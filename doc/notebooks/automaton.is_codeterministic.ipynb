# _automaton_.is_codeterministic

Whether the automaton is codeterministic (its transposition is deterministic).
    

See also:
- [automaton.is_deterministic](automaton.is_deterministic.ipynb)
- [automaton.determinize](automaton.determinize.ipynb)
- [automaton.codeterminize](automaton.codeterminize.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton -s a
context = "lal(ab), b"
$ -> 0
0 -> 1 a
0 -> 2 b
1 -> 3 b
2 -> 3 b
3 -> $
```

```python
a.is_codeterministic()
```

```python
a.transpose().is_deterministic()
```
