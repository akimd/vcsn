# _automaton_.is_useless

Whether the automaton is _useless_, i.e., whether it has successful computations. Morally, should be equivalent to whether it accepts no words, but see below.

This is equivalent to testing whether the trim part is empty.

Caveat:
- an non-useless automaton might still accept no words (see below).

See also:
- [_automaton_.is_empty](automaton.is_empty.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton -s a
0 -> 0 a, b
0 -> 1 a
1 -> 1 a, b
```

This automaton is not empty, but totally useless:

```python
a.is_empty()
```

```python
a.is_useless()
```

If we trim it, then it is empty (which is why it seems that there is no result):

```python
a.trim()
```

```python
a.trim().is_empty()
```

### Caveat

On some specific structures, `automaton.is_useless` returns false although the automaton accepts no words.

```python
%%automaton -s a
context = "lal_char, z"
$ -> 0
0 -> 1 a
0 -> 2 <-1>a
1 -> $
2 -> $
```

```python
a.is_useless()
```

```python
a.shortest(10)
```
