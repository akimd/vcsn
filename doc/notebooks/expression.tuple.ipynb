# `expression | exp`

Build the tuple of expressions, ie the $k$-tape expression whose behavior is the Cartesian product of the behaviors of the input expressions.

See also:
- [automaton.tuple](automaton.tuple.ipynb)

## Examples

```python
import vcsn
exp = vcsn.context('lan, b').expression
e1 = exp('abc')
e1
```

```python
e2 = exp('x*')
e2
```

```python
e1 | e2
```

```python
(e1 | e2).automaton()
```

```python
e1.automaton() | e2.automaton()
```
