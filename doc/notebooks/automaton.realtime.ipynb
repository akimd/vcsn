_automaton_.realtime
======

Compute the realtime automaton, i.e. the equivalent automaton in which transitions are labeled with only *letters* (as viewed by the labelset), without spontaneous transitions.

Calling `realtime` is equivalent to calling `letterize` then `proper`.

Preconditions:

- The automaton `is_valid`

Postconditions:

- The automaton `is_letterized`

- The automaton `is_proper`

See also:

- [automaton.letterize](automaton.letterize.ipynb)

- [automaton.is_letterized](automaton.is_letterized.ipynb)

- [automaton.is_realtime](automaton.is_realtime.ipynb)

- [automaton.is_valid](automaton.is_valid.ipynb)

- [automaton.proper](automaton.proper.ipynb)

Examples
-----

```python
import vcsn
```

```python
law = vcsn.context("law_char, b")
```

```python
a = law.expression("(ab)*+(cd+ef)*").thompson()
a
```

```python
a.is_valid()
```

```python
a.is_letterized()
```

```python
a.is_proper()
```

```python
r = a.realtime()
r
```

```python
r.context()
```

```python
r.is_letterized()
```

```python
r.is_proper()
```

```python
a.letterize().proper()
```
