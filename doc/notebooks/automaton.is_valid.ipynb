# _`automaton`_`.is_valid`

A careful analysis of automata with spontaneous transitions shows that in some case, spontaneous-cycles may result in automata with an undefined behavior.  They are called _invalid_.

Preconditions:
- None

See also:
- [automaton.proper](automaton.proper.ipynb)
- [expression.is_valid](expression.is_valid.ipynb)

References:
- [lombardy.2013.ijac](References.ipynb#lombardy.2013.ijac)

## Examples

```python
import vcsn
```

The following examples are taken from [lombardy.2013.ijac](References.ipynb#lombardy.2013.ijac).

#### Automaton $\mathcal{Q}_3$ (Fig. 2) (in $\mathbb{Q}$)

The following automaton is invalid.

```python
%%automaton q3
context = "lan_char, q"
$ -> 0
0 -> 0 <-1/2>\e
0 -> 1  <1/2>\e
1 -> 1 <-1/2>\e
1 -> 0  <1/2>\e
0 -> $
```

```python
q3.is_valid()
```

#### Automaton $\mathcal{Q}_4$ (Fig. 3) (in $\mathbb{Q}$)

The following one, however, is valid.  Spontaneous transitions can be eliminated.

```python
%%automaton q4
context = "lan_char, q"
$ -> 0
0 -> 1 <1/2>\e, a
1 -> 0 <-1>\e, b
1 -> $ <2>
```

```python
q4.proper()
```

#### An Invalid Thompson Automaton (Fig. 5)

Sadly enough, the (weighted) Thompson construction may build invalid automata from valid expressions.

```python
e = vcsn.context('lal_char, q').expression('(a*+<-1>b*)*')
e.is_valid()
```

```python
t = e.thompson()
t
```

```python
t.is_valid()
```
