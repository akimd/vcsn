# `automaton | aut`

The (accessible part of the) "tuple" of automata, ie the $k$-tape automaton whose behavior is the Cartesian product of the behaviors of the input automata.

See also:
- [expansion.tuple](expansion.tuple.ipynb)
- [expression.tuple](expression.tuple.ipynb)
- [polynomial.tuple](polynomial.tuple.ipynb)


## Examples

```python
import vcsn
exp = vcsn.context('lal, q').expression
e1 = exp('abc')
a1 = e1.automaton()
a1
```

```python
e2 = exp('(<2>x)*')
a2 = e2.automaton()
a2
```

```python
a1 | a2
```

```python
(e1 | e2).automaton()
```
