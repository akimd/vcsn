# automaton.conjugate
Create the conjugate of an automaton.
The conjugate of $\mathcal{A}$ is defined as the automaton recognizing the language $\{vu \mid u, v \in \Sigma^*\text{ and }uv\text{ recognized by }\mathcal{A}\}$.

Preconditions:
- labelset is nullable

Postconditions:
- None

# Examples

```python
import vcsn
```

```python
ctx = vcsn.context('lan_char, b')
a = ctx.expression('ab*').automaton()
a
```

```python
a.shortest(10)
```

```python
conj = a.conjugate()
conj
```

```python
conj.shortest(10)
```
