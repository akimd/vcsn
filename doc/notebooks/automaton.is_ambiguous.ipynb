# _automaton_.is_ambigous

Whether the automaton is ambiguous.

Preconditions:
- the automaton is free.

Postconditions:
- [_automaton_.ambiguous_word](automaton.ambiguous_word.ipynb) does not raise.

See also:
- [_automaton_.ambiguous_word](automaton.ambiguous_word.ipynb)
- [_automaton_.is_cycle_ambiguous](automaton.is_cycle_ambiguous.ipynb)

## Examples

```python
import vcsn
a = vcsn.B.expression('abc').standard()
a
```

```python
a.is_ambiguous()
```

```python
a + a
```

```python
(a+a).is_ambiguous()
```

```python
a = vcsn.B.expression('(a+aa)*').automaton()
a
```

```python
a.is_ambiguous()
```
