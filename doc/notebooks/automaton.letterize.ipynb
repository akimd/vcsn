# automaton.letterize()

Split the transition labels into letters.

Preconditions:
- None

Postconditions:
- the output transitions are *atomic*, meaning that they are composed of one *letter* (as viewed by the labelset). The labelset is adapted to provide this guarantee.

See also:
- [automaton.is_letterized](automaton.is_letterized.ipynb)
- [automaton.is_realtime](automaton.is_realtime.ipynb)
- [automaton.realtime](automaton.realtime)

## Examples

```python
import vcsn
ctx = vcsn.context("law, b")
ctx
```

```python
a = ctx.expression("(cat)+(dog)*").standard()
a
```

```python
al = a.letterize()
al
```

```python
al.context()
```

Note that here, the transitions are no longer *words* (*law*) but instead *letters* (*lal*).

```python
ctx_t = vcsn.context("lat<law, law, lal>, z")
```

```python
b = ctx_t.expression(r"(<2>'cat,dog,c')*+'\e,b,d'").derived_term()
b
```

```python
bl = b.letterize()
bl
```

```python
bl.context()
```

Here, all three labelsets have been converted to letters or the empty word (`lal`).
