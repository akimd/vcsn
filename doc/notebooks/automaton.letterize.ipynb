# automaton.letterize()

Split the transition labels into letters.

Preconditions:
- None

Postconditions:
- the output transitions are *letterized*, meaning that they are composed of one *letter*. The labelset is adapted to provide this guarantee.

See also:
- [automaton.is_letterized](automaton.is_letterized.ipynb)
- [automaton.is_realtime](automaton.is_realtime.ipynb)
- [automaton.realtime](automaton.realtime)

## Examples

```python
import vcsn
ctx = vcsn.context("[...]* -> b")
ctx
```

```python
a = ctx.expression("(cat)+(dog)*").standard()
a
```

```python
al = a.letterize()
al
```

```python
al.context()
```

The transitions are no longer labeled with words (`[...]*`) but with letters (`[...]?`).

```python
ctx = vcsn.context("[...]* x [...]* x [...] -> Z")
b = ctx.expression(r"(<2>'cat,dog,c')*+'\e,b,d'").derived_term()
b
```

```python
bl = b.letterize()
bl
```

```python
bl.context()
```

Here, all three labelsets have been converted to letters or the empty word (`[...]`).
