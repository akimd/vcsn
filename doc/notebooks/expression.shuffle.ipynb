# _`expression`_`.shuffle(exp)`

The (accessible part of the) shuffle product of expressions.

See also:
- [expression.conjunction](expression.conjunction.ipynb)
- [expression.infiltrate](expression.infiltrate.ipynb)
- [automaton.shuffle](automaton.shuffle.ipynb)

## Examples

```python
import vcsn
```

### Boolean Automata

The shuffle product of automata computes the shuffling of their languages: all the possible interleavings.

```python
a = vcsn.B.expression('abc')
a
```

```python
b = vcsn.B.expression('xyz')
b
```

```python
shuff = a.shuffle(b)
shuff
```

```python
shuff.automaton()
```

### Weighted expression

In the case of weighted expressions, weights are "kept" with the letters.

```python
a = vcsn.Z.expression('<2>a<3>b', 'none')
b = vcsn.Z.expression('<4>x<-2>y', 'none')
a.shuffle(b).automaton()
```

### Associativity
The operator is associative: `a.shuffle(b).shuffle(c)` is the same as `a.shuffle(b.shuffle(c))`:

```python
exp = lambda e: vcsn.Z.expression(e)
x = exp('<2>a')
y = exp('<3>a')
z = exp('<4>a')
xyz = x.shuffle(y).shuffle(z)
yzx = x.shuffle(y.shuffle(z))
```
