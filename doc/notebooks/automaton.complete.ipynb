# _automaton_.complete

Make an automaton complete, i.e., there is an outgoing transition from every state for every labels.

Precondition:
- the automaton is free

Postcondition:
- Result.is_complete()

See also:
- [_automaton_.is_complete](automaton.is_complete.ipynb)

## Examples

```python
import vcsn
z = vcsn.context('lal(abc), z')
```

```python
a = z.expression('(<2>a<3>b)*').standard()
a
```

```python
a.is_complete()
```

```python
a.complete()
```

```python
a.complete().is_complete()
```
