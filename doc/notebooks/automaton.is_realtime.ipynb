# _automaton_.is_realtime

Check whether the automaton is _realtime_, i.e. it has only *letter* transitions (as viewed the *labelset*), so no words or spontaneous transitions.

It is equivalent to being both `is_letterized` and `is_proper`.

See also:
- [automaton.is_letterized](automaton.is_letterized.ipynb)
- [automaton.is_proper](automaton.is_proper.ipynb)
- [automaton.realtime](automaton.realtime.ipynb)

Examples
-----

```python
import vcsn
ctx = vcsn.context("law_char, b")
```

```python
a1 = ctx.expression("abc").standard()
a1
```

```python
a1.is_realtime()
```

Here, a1 has a _word_ transition, so it is not _letterized_.

```python
a2 = ctx.expression("a*+b(c+d)").thompson()
a2
```

```python
a2.is_realtime()
```

Here, a2 has _spontaneous transitions_, so it is not _proper_.

```python
a3 = ctx.expression("a*+b(c+d)").standard()
a3
```

```python
a3.is_realtime()
```
