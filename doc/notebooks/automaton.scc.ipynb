# _automaton_.scc(algo = "auto")

Compute the strongly-connected components and decorate the input automaton with them.

The algorithm can be:
- "auto": same as "tarjan_iterative".
- "tarjan_iterative": the algorithm from Tarjan implemented without recursion.  Fast and robust to deep automata.
- "tarjan_recursive": the algorithm from Tarjan implemented with recursion.  Faster than "tarjan_iterative", but might explode on very deep automata.
- "kosaraju": Using the algorithm of Kosaraju.  Slower.

## Examples

```python
import vcsn
z = vcsn.context('lal_char(abc), z')
```

```python
a = z.expression('(<2>a<3>b)*').standard()
a
```

This automaton has two strongly connected components:
+ Component 1: {1, 3} 
+ Component 2: {0}

```python
c = a.scc()
c
```

```python
c.is_isomorphic(a)
```

### _`scc`_.`component`(_`num`_)

We can view with only one component:

```python
c.component(2)
```

### _`scc`_.`condense`()

Or view the automaton as condensation:

```python
c.condense()
```

### _`scc`_.`num_components`()

View number of strongly connected components:

```python
c.num_components()
```

```python
a = z.expression("(abc)*{2}").standard()
a
```

This automaton has 3 components.

```python
c = a.scc("tarjan_recursive")
c
```

```python
c.is_isomorphic(a)
```

```python
c.component(1)
```

```python
c.condense()
```

```python
c.num_components()
```
