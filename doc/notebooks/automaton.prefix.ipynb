# _automaton_.prefix

An automaton that accepts all the prefixes of the words accepted by the previous automaton.

Preconditions:
- None

See also:
- [_automaton_.factor](automaton.factor.ipynb)
- [_automaton_.subword](automaton.subword.ipynb)
- [_automaton_.suffix](automaton.suffix.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton -s a
context = "lal_char(abc), q"
$ -> 0 <2>
0 -> 1 <3>a
1 -> 1 <4>b
1 -> 2 <5>c
2 -> 3 <6>a
3 -> 4 <7>b
3 -> 5 <8>c
4 -> $ <9>
6 -> 0 <10>c
```

```python
a.prefix()
```
