# _automaton_.ambiguous_word

Return an ambiguous word, or raise if the automaton is unambiguous.

Preconditions:
- the labelset is free

Postconditions:
- the Result is accepted

See also:
- [_automaton_.is_ambiguous](automaton.is_ambiguous.ipynb)

## Examples

```python
import sys
import vcsn
a = vcsn.B.expression('abc').standard()
a
```

```python
a.is_ambiguous()
```

```python
try:
    a.ambiguous_word()
except RuntimeError as err:
    print("error:", err, file=sys.stderr)
```

```python
a = vcsn.B.expression('in (xxyy + xx + yy)* out').automaton()
a
```

```python
a.ambiguous_word()
```
