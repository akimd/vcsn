# _automaton_.is_accessible

Whether all its states are accessible, i.e., all its states can be reached from an initial state.

Preconditions:
- None

See also:
- [_automaton_.accessible](automaton.accessible.ipynb)
- [_automaton_.is_coaccessible](automaton.is_coaccessible.ipynb)
- [_automaton_.trim](automaton.trim.ipynb)

## Examples

```python
import vcsn
```

The following automaton has states that cannot be reached from the initial(s) states:

```python
%%automaton a
context = "[abc]"
$ -> 0
0 -> 1 a
1 -> $
2 -> 0 a
1 -> 3 a
```

```python
a.is_accessible()
```

Calling `accessible` returns a copy of the automaton without non-accessible states:

```python
a.accessible()
```

```python
a.accessible().is_accessible()
```
