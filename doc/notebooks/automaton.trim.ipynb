# _automaton_.trim

Create a new automaton from the _useful_ part of the input, i.e., the subautomaton whose states both accessible (can be reached from a initial state) and coaccessible (can reach a final state).

Preconditions:
- None

Postconditions:
- Result.is_accessible
- Result.is_coaccessible
- Result.is_trim

See also:
- [_automaton_.is_coaccessible](automaton.is_accessible.ipynb)
- [_automaton_.accessible](automaton.accessible.ipynb)
- [_automaton_.trim](automaton.trim.ipynb)

## Examples

```python
import vcsn
```

The following automaton has states that cannot be reach any final(s) states:

```python
%%automaton -s a
context = "lal(abc), b"
$ -> 0
0 -> 1 a
1 -> $
2 -> 0 a
1 -> 3 a
```

```python
a.is_trim()
```

Calling `trim` returns the same automaton, but without any useless states:

```python
a.trim()
```

```python
a.trim().is_trim()
```
