# _automaton_.subword

An automaton that accepts all the suffixes of the subwords accepted by the previous automaton.

Preconditions:
- None

See also:
- [_automaton_.factor](automaton.factor.ipynb)
- [_automaton_.prefix](automaton.prefix.ipynb)
- [_automaton_.suffix](automaton.suffix.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton -s a
context = "lal_char(abc), q"
$ -> 0 <2>
0 -> 1 <3>a
1 -> 1 <4>b
1 -> 2 <5>c
2 -> 3 <6>a
3 -> 4 <7>b
3 -> 5 <8>c
4 -> $ <9>
6 -> 0 <10>c
```

```python
a.subword()
```
