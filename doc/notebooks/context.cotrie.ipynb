# _context_.cotrie(_`file`_)

Generate a "cotrie" automaton from a finite series, given as a file of (weighted) words.

Postconditions:
- `Result.is_codeterministic()`

See also:
- [context.trie](polynomial.trie.ipynb)
- [polynomial.cotrie](polynomial.cotrie.ipynb)
- [polynomial.trie](polynomial.trie.ipynb)

## Examples

```python
import vcsn
```

### Words (finite language)

```python
%%file words
hello
world
hell
word
```

```python
vcsn.B.cotrie('words')
```

### Weighted words (finite series)

```python
%%file numbers
one
<2>two
<3>three
<4>four
<1/2>half
```

```python
vcsn.Q.cotrie('numbers')
```

### Tuples of words

```python
%%file translator
<1>one|un
<2>two|deux
<3>three|trois
<4>four|quatre
<14>forteen|quatorze
<40>forty|quarante
```

```python
vcsn.context('lat<law_char, law_char>, q').cotrie('translator')
```
