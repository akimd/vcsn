# _automaton_.strip

Remove all the decorations from an automaton, i.e., remove state names.

## Examples

```python
import vcsn
```

```python
a = vcsn.B.expression('ab*').automaton()
a
```

Stripping a plain automaton, or stripping twice, is allowed.

```python
a.strip()
```

```python
a.strip().strip()
```

Operations like [determinize](automaton.determinize.ipynb) introduce decorations:

```python
a = vcsn.context('[ab]').de_bruijn(1).determinize()
a
```

```python
a.strip()
```

```python
a.type()
```

```python
a.strip().type()
```

The [derived_term](expression.derived_term.ipynb) of an expression also uses decorations:

```python
a = vcsn.Z.expression('<3>ab*c{3}').derived_term()
a
```

```python
a.strip()
```
