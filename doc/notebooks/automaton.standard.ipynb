# _automaton_.standard

Turn an automaton into a standard automaton.

Preconditions:
- None

Postconditions:
- `result.is_standard()`

See also:
- [_automaton_.is_standard](automaton.is_standard.ipynb)
- [_expression_.derived_term](expression.derived_term.ipynb)
- [_expression_.standard](expression.standard.ipynb)

## Examples

There is a single initial state.

```python
import vcsn
ctx = vcsn.context('lal_char, z')
std = lambda e: ctx.expression(e).standard()
a = std('a').add(std('b'), 'general')
a
```

```python
a = a.standard()
a
```

The initial state has 1 as (initial) weight.

```python
%%automaton -s a
context = "lal_char, z"
$ -> 0 <2>
0 -> 1 a
0 -> 2 b
1 -> $
2 -> $
```

```python
a.standard()
```

There is no incoming transition on the initial state.

```python
%%automaton -s a
context = "lal_char, z"
$ -> 0 <2>
0 -> $ <2>
0 -> 1 a
1 -> 0 b
```

```python
a.standard()
```
