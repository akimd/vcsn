# _ratexp_.standard

Generate the standard automaton from an expression.

Postcondition:
- `Result.is_standard()`

Caveats:
- it is not guaranteed that `Result.is_valid()`

See also:
- [_automaton_.standard](automaton.standard.ipynb)
- [_automaton_.is_standard](automaton.is_standard.ipynb)

## Examples

```python
import vcsn
q = vcsn.context('lal_char(abc)_q')
```

```python
q.ratexp('(<1/6>a*+<1/3>b*)*').standard()
```
