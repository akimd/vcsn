<!--I graph operations -->
A number of functions on automata deal with their graph structure, irrespective of labels.  They are defined for any context.

$reverse$
=======
<!--I $reverse$ -->
$reverse(a)$: return a copy of $a$ where every edge is reversed, and final states are made initial and initial states are made final.

**The function is not implemented**. It is just convenient to *specify* "dual" functions in this chapter.
It differs from $transpose$ as it has no effect on labels.


<!-- `accessible`, `coaccessible`, `trim` -->
$accessible$, $coaccessible$, $trim$
==============================

<!--I $accessible$ -->
These functions return a sub-automaton of the given automaton, without decorations.  They are all non-destructive.
Type is $(L \to W) \to (L \to W)$ in every case.

$accessible$
------------
$accessible(a)$: return a copy of the accessible part of the automaton $a$.

The description of the function is the specification.
The function is implemented by a traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `accessible(`$a$`)`

* *method*:
  $a$`.accessible()`

* *Dyn*:
  `accessible(`$a$`)`

Output decoration:
  *none*
  

$coaccessible$
------------
<!--I $coaccessible$ -->
$coaccessible(a)$: return a copy of the co-accessible part of the automaton $a$.

Specification:
$coaccessible(a) \equiv reverse(accessible(reverse(a)))$.  
It is realised by a traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `coaccessible(`$a$`)`

* *method*:
  $a$`.coaccessible()`

* *Dyn*:
  `coaccessible(`$a$`)`

Output decoration:
  *none*
  
$trim$
------------
<!--I $trim$ -->
$trim(a)$: return a copy of the trim part of the automaton $a$.

Specification:
$trim(a) \equiv coaccessible(accessible(a))$.  
It is realised by a double traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `trim(`$a$`)`

* *method*:
  $a$`.trim()`

* *Dyn*:
  `trim(`$a$`)`

Output decoration:
  *none*

$is\_accessible$, $is\_coaccessible$, $is\_trim$, $is\_empty$
==============================
<!--I predicate -->
These non-destructive predicates check a property of the given automaton and return a boolean as result.
Type is $(L \to W) \to \mathbb{B}$ in every case.

$is\_accessible$
------------
<!--I $is\_accessible$ -->
$is\_accessible(a)$: return `true` iff the automaton $a$ is accessible.

Specification: $is\_accessible(a)$ = ($accessible(a) \equiv a$)

Call syntax:

* *functional*:
  `is_accessible(`$a$`)`

* *method*:
  $a$`.is_accessible()`

* *Dyn*:
  `is_accessible(`$a$`)`

$is\_coaccessible$
------------
<!--I $is\_coaccessible$ -->
$is\_coaccessible(a)$: return `true` iff the automaton $a$ is co-accessible.

Specification: $is\_coaccessible(a)$ = ($coaccessible(a) \equiv a$)

Call syntax:

* *functional*:
  `is_coaccessible(`$a$`)`

* *method*:
  $a$`.is_coaccessible()`

* *Dyn*:
  `is_coaccessible(`$a$`)`

$is\_trim$
------------
<!--I $is\_trim$ -->
$is\_trim(a)$: return `true` iff the automaton $a$ is trim.

Specification: $is\_trim(a)$ = ($trim(a) \equiv a$)

Call syntax:

* *functional*:
  `is_trim(`$a$`)`

* *method*:
  $a$`.is_trim()`

* *Dyn*:
  `is_trim(`$a$`)`

$is\_empty$
------------
<!--I $is\_empty$ -->
$is\_empty(a)$: return `true` iff the automaton $a$ is empty.

Specification: same as description.

Call syntax:

* *functional*:
  `is_empty(`$a$`)`

* *method*:
  $a$`.is_empty()`

* *Dyn*:
  `is_empty(`$a$`)`
