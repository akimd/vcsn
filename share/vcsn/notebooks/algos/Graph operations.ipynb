A number of functions on automata deal with their graph structure, irrespective of labels.  They are defined for any context.

$reverse$
=======

$reverse(a)$: return a copy of $a$ where every edge is reversed, and final states are made initial and initial states are made final.

**The function is not implemented**. It is just convenient to *specify* "dual" functions in this chapter.
It differs from $transpose$ as it has no effect on labels.

$accessible$, $coaccessible$, $trim$
==============================

These functions return a sub-automaton of the given automaton, without decorations.  They are all non-destructive.

$accessible$
------------
$accessible(a)$: return a copy of the accessible part of the automaton $a$.

The description of the function is the specification.
The function is implemented by a traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `accessible(`$a$`)`

* *method*:
  $a$`.accessible()`

* *Dyn*:
  `accessible(`$a$`)`

Output decoration:
  *none*
  

$coaccessible$
------------
$coaccessible(a)$: return a copy of the co-accessible part of the automaton $a$.

Specification:
$coaccessible(a) \equiv reverse(accessible(reverse(a)))$.  
It is realised by a traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `coaccessible(`$a$`)`

* *method*:
  $a$`.coaccessible()`

* *Dyn*:
  `coaccessible(`$a$`)`

Output decoration:
  *none*
  
$trim$
------------
$trim(a)$: return a copy of the trim part of the automaton $a$.

Specification:
$trim(a) \equiv coaccessible(accessible(a))$.  
It is realised by a double traversal of the underlying graph of $a$.


Call syntax:

* *functional*:
  `trim(`$a$`)`

* *method*:
  $a$`.trim()`

* *Dyn*:
  `trim(`$a$`)`

Output decoration:
  *none*

$is\_accessible$, $is\_coaccessible$, $is\_trim$
==============================

These non-destructive functions check a property of the given automaton and return a boolean as result.

$is\_accessible$
------------
$is\_accessible(a)$: return true iff the automaton $a$ is accessible.

Specification: $is\_accessible(a)$ = ($accessible(a) \equiv a$)

Call syntax:

* *functional*:
  `is_accessible(`$a$`)`

* *method*:
  $a$`.is_accessible()`

* *Dyn*:
  `is_accessible(`$a$`)`

$is\_coaccessible$
------------
$is\_coaccessible(a)$: return true iff the automaton $a$ is co-accessible.

Specification: $is\_coaccessible(a)$ = ($coaccessible(a) \equiv a$)

Call syntax:

* *functional*:
  `is_coaccessible(`$a$`)`

* *method*:
  $a$`.is_coaccessible()`

* *Dyn*:
  `is_coaccessible(`$a$`)`

$is\_trim$
------------
$is\_trim(a)$: return true iff the automaton $a$ is trim.

Specification: $is\_trim(a)$ = ($trim(a) \equiv a$)

Call syntax:

* *functional*:
  `is_trim(`$a$`)`

* *method*:
  $a$`.is_trim()`

* *Dyn*:
  `is_trim(`$a$`)`

$is\_empty$, $is\_useless$
==============================

$is\_empty$
------------
$is\_empty(a)$: return true iff the automaton $a$ is empty.

Specification: same as description.

Call syntax:

* *functional*:
  `is_useless(`$a$`)`

* *method*:
  $a$`.is_useless()`

* *Dyn*:
  `is_useless(`$a$`)`


$is\_useless$
------------
$is\_useless(a)$: return `true` iff $a$ has successful computations.

Specification: $is\_useless(a) = is\_accessible(a) \wedge is\_coaccessible(a)$.

Call syntax:

* *functional*:
  `is_useless(`$a$`)`

* *method*:
  $a$`.is_useless()`

* *Dyn*:
  `is_useless(`$a$`)`

$is\_useless$ is a *graph function* and tests whether there
are *successful computations* in the automaton, i.e. a
sequence of co-terminal transitions, the first one beginning in an
`initial state', the last one ending in a `final state'.
By definition, or by the way automata are specified in Vaucanson,
each of these transitions have a non-zero label.
This does not imply that the label of the computation itself is
different from zero, nor that the behaviour of the automaton is
different from zero.

For instance, the behaviour of the following automaton is the null series; nonetheless it is not useless.

```python
from vcsn import *
a = context('lal_char(a-d)_z').ratexp('a+<-1>b').derived_term().strip()
a
```

```python
a.is_empty()
```
