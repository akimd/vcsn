# _automaton_.cominimize(string algo = "weighted")

Cominimizes a automaton

Calling aut.cominimize() is equivalent to call aut.transpose().minimize().transpose() 


See also:
- [automaton.minimize](automaton.minimize.ipynb)
- [automaton.transpose](automaton.transpose.ipynb)


## Examples

```python
import vcsn
a = vcsn.automaton('''
  vcsn_context = "lal_char(abc)_z"
  $ -> 0
  $ -> 1
  0 -> 2 <4>a
  1 -> 3 a
  2 -> 2 a
  3 -> 3 a
  2 -> 4 <3>a
  3 -> 4 <4>a
  4 -> $
''', 'daut')
a
```

```python
a1 = a.cominimize()
a1
```

```python
a2 = a.transpose().minimize().transpose()
a2
```

```python
a1.is_equivalent(a2)
```
