# _automaton_.codeterminize

Computes the transposition of determinization of transpositive automaton.

Preconditions:
- its labelset is free
- its weightset features a division operator (which is the case for $\mathbb{B}$).

Postconditions:
- the result is deterministic
- the result is equivalent to the input.transpose().determinize().transpose()

Caveats:
- might not terminate if the weightset is not $\mathbb{B}$

See also:
- [automaton.determinize](automaton.determinize.ipynb)
- [automaton.transpose](automaton.transpose.ipynb)
- [automaton.is_deterministic](automaton.is_deterministic.ipynb)

## Examples

```python
import vcsn
```

```python
lady3 = vcsn.context('lal_char(abc)_b').ladybird(3)
lady3
```

```python
lady3cd = lady3.codeterminize()
lady3cd
```

The resulting automaton has states labeled with subsets of the input automaton set of states.

```python
lady3cd2 = lady3.transpose().determinize().transpose()
lady3cd2
```

```python
lady3cd.enumerate(4)
```

```python
lady3cd2.enumerate(4)
```

Two automata are equivalent, for example:

```python
lady3cd.enumerate(4) == lady3cd2.enumerate(4)
```
