# _automaton_.filter(_states_)

Return a subautomaton such that their states are in the input states set.

Postcondition:
- The result automaton is subautomaton of input automaton.
- The states of automaton are in the input states set.

See also:
- [_automaton_.accessible](automaton.accessible.ipynb)
- [_automaton_.coaccessible](automaton.coaccessible.ipynb)
- [_automaton_.trim](automaton.trim.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton aut
context = "lal_char(a)_b"
0 -> 1 a
1 -> 0 a
0 -> 4 a
1 -> $
1 -> 2 a
$ -> 0
3 -> 4 a
4 -> 0 a
4 -> 5 a
```

When states set $\{0, 1, 2 , 3\}$ is subset of set of states of automaton $\{0, 1, 2 , 3, 4, 5\}$

```python
aut.filter([0, 1, 2, 3])
```

When states set $\{0, 1, 7\}$, the result set is intersection of this set with set of state of automaton: $\{0, 1\}$

```python
aut.filter([0, 1, 7])
```

We can use `filter` of `filter` as below:

```python
aut.filter([0, 1, 2, 3, 4]).filter([0, 1, 3, 5])
```

We can filter a transposed automaton:

```python
trans = aut.transpose()
trans
```

```python
trans.filter([0, 1, 2, 4])
```
