# _automaton_.scc(algo = "auto")

Make a automaton that each state has name corresponding                                                                                                              
$[component.state]$, for example $[1.2]$ means that state $2$ has                                                                                                           
component number $1$.

The algorithm can be:
+ "auto" or "tarjan_iterative": By defaut, using algorithm tarjan iterative.
+ "tarjan_recursive": Using algorithm tarjan recursive.
+ "kosaraju": Using algorithm kosaraju.

Postcondition:
- Result automaton and input automaton are isomorphic.


## Examples

```python
import vcsn
z = vcsn.context('lal_char(abc), z')
```

```python
a = z.expression('(<2>a<3>b)*').standard()
a
```

This automaton has two strongly connected components:
+ Component 1: {1, 3} 
+ Component 2: {0}

```python
c = a.scc()
c
```

```python
c.is_isomorphic(a)
```

We can view with only one component:

```python
c.component(2)
```

Or view the automaton as condensation:

```python
c.condense()
```

View number of strongly connected components:

```python
c.num_components()
```

```python
a = z.expression("(abc)*{2}").standard()
a
```

This automaton has 3 components.

```python
c = a.scc("tarjan_recursive")
c
```

```python
c.is_isomorphic(a)
```

```python
c.component(1)
```

```python
c.condense()
```

```python
c.num_components()
```
