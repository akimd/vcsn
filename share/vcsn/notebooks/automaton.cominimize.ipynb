# _automaton_.cominimize(algo = "auto")

Cominimize an automaton.  Calling `aut.cominimize()` is equivalent `aut.transpose().minimize().transpose()`.

Arguments:
- `algo` the algorithm to perform the minimization (see minimization for details).

See also:
- [automaton.minimize](automaton.minimize.ipynb)
- [automaton.transpose](automaton.transpose.ipynb)

## Examples

```python
import vcsn
```

```python
%%automaton a
context = "lal_char(abc), z"
$ -> 0
$ -> 1
0 -> 2 <4>a
1 -> 3 a
2 -> 2 a
3 -> 3 a
2 -> 4 <3>a
3 -> 4 <4>a
4 -> $
```

```python
a1 = a.cominimize()
a1
```

```python
a2 = a.transpose().minimize().transpose()
a2
```

```python
a1.is_equivalent(a2)
```
