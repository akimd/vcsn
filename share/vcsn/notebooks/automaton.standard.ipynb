# _automaton_.standard

Turns an automaton into a standard automaton.

Preconditions:
- None

Postconditions:
- `result.is_standard()`

See also:
- [_automaton_.is_standard](automaton.is_standard.ipynb)
- [_ratexp_.derived_term](ratexp.derived_term.ipynb)
- [_ratexp_.standard](ratexp.standard.ipynb)
- [_ratexp_.thompson](ratexp.thompson.ipynb)

## Examples

There is a single initial state.

```python
import vcsn
b = vcsn.context('lal_char(abc), b')
a = b.ratexp('a').standard() | b.ratexp('b').standard()
a
```

```python
a.standard()
```

The initial state has 1 as (initial) weight.

```python
a = vcsn.automaton('''
digraph {
  vcsn_context = "lal_char(abc), z"
   I -> 0 [label = "<2>"]
   0 -> 1 [label = "a"]
   1 -> F
}''')
a
```

```python
a.standard()
```

There is no incoming transition on the initial state.

```python
a = vcsn.automaton('''
digraph {
  vcsn_context = "lal_char(abc), z"
   I -> 0 -> F [label = "<2>"]
   0 -> 1 [label = "a"]
   1 -> 0 [label = "b"]
}''')
a
```

```python
a.standard()
```
