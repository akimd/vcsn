# Basic Operations on Automata

```python
import vcsn
```

```python
lb = vcsn.context('lal_char(a-z)_b')
wb = vcsn.context('law_char(a-z)_b')
```

## Transpose

```python
lba = lb.ratexp('(ab)(cd)(ef)').linear()
```

To transpose an automaton, use `automaton.tranpose`:

```python
lba.transpose()
```

Transposing a transposed automaton return the original automaton.

```python
lba.transpose().transpose()
```

```python
lba == lba.transpose().transpose()
```

Of course labels and weights are transposed too when it applies.

```python
wb.ratexp('(abc)(def)').linear()
```
