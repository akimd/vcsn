```python
from IPython.display import display
from IPython.html import widgets
from IPython.utils import traitlets

class Automaton(widgets.DOMWidget):
    from IPython.display import Javascript

    # the name of the Backbone.View subclass to be used
    _view_name = traitlets.Unicode('AutomatonView', sync=True)
    nodes = traitlets.List(sync=True)
    links = traitlets.List(sync=True)
    lastNodeId = traitlets.CInt(sync=True)
```

```python
%%javascript
require.config({paths: {d3: "//d3js.org/d3.v3.min"}});
```

```python
%%javascript
require(['widgets/js/widget','d3','jquery','underscore'], function(WidgetManager, d3, $,  _){
    var AutomatonView = IPython.DOMWidgetView.extend({
             
        // RENDER THE VIEW
        render: function(){
            
            // Here we define a variable with this so we can call from everywhere
            
            // Attributs of the SVG
            this.width =960, this.height = 500;
            
            // Render the svg
            this.svg = d3.select(this.el).append('svg')
                .attr('width', this.width)
                .attr('height', this.height)
                .style('cursor', 'crosshair')
                .style('userSelect' , "none")
                .style('webkiUserSelect' , "none")
                .style('MozUserSelect' , "none")
                .style('oUserSelect', 'none')
                .style('msUserSelect', 'none')
                .style('draggable', 'false');
            
            // Define mouse and keyboard variables
            this.selected_node = null,
            this.selected_link = null,
            this.mousedown_link = null,
            this.mousedown_node = null,
            this.mouseup_node = null;
            this.lastKeyDown = -1;
            
            // Initialize datas
            this.nodes = this.model.get('nodes');
            this.links = this.model.get('links');
            this.lastNodeId = this.nodes[this.nodes.length - 1].id;
            
            // Define path and circle here to avoid recreation of svg in every update
            this.path = this.svg.append('svg:g').selectAll('g'); 
            this.circle = this.svg.append('svg:g').selectAll('g');
            
            // Lines displayed when draggind new nodes
            this.drag_line = this.svg.append('svg:path')
                .style('fill', 'none')
                .style('stroke', '#000')
                .style('stroke-width', '1.75px')
                .attr('class', 'link dragline hidden')
                .attr('d', 'M0,0L0,0');
            
            // Arrow marker for links
            this.svg.append('svg:defs').append('svg:marker')
                .attr('id', 'end-arrow')
                .attr('viewBox', '0 -5 10 10')
                .attr('refX', 10)
                .attr('markerWidth', 6)
                .attr('markerHeight', 6)
                .attr('orient', 'auto')
              .append('svg:path')
                .attr('d', 'M0,-5L10,0L0,5')
                .attr('fill', '#000');

            // Variable using in keyboard functions
            this.overtext = false;
            
            // Closure 
            var that = this;

            // The force layout for physical simulation
            this.force = d3.layout.force()
                .nodes(this.nodes)
                .links(this.links)
                .size([this.width, this.height])
                .linkDistance(function(d) {
                    if(d.short == true) {return 75;}
                    else {return 200;}
                })
                .charge(function(d) {
                    if(d.hidden == true) {return -100;}
                    else {return -750;}
                })
                .on('tick', function(d,i){ that.stepForce(this, d, i);})

            // Listen for mouse events on the whole document
            this.svg.on('mousedown', function(d,i){ that.mousedown(this, d, i);})
                .on('mousemove', function(d,i){ that.mousemove(this, d, i);})
                .on('mouseup', function(d,i){ that.mouseup(this, d, i);});
            
            this.update();
        },
        
        // Set position of each elements for the tick event
        stepForce: function(doc, d, i) {
            // draw links with proper padding from node centers
            this.path.selectAll('path').attr('d', function(d) {
                if(d.target != d.source) {
                    var deltaX = d.target.x - d.source.x ,
                        deltaY = d.target.y - d.source.y,
                        dist = Math.sqrt(deltaX * deltaX + deltaY * deltaY),
                        normX = deltaX / dist,
                        normY = deltaY / dist,
                        sourcePadding = 12,
                        targetPadding = 17,
                        sourceX = d.source.x + (sourcePadding * normX),
                        sourceY = d.source.y + (sourcePadding * normY),
                        targetX = d.target.x - (targetPadding * normX),
                        targetY = d.target.y - (targetPadding * normY);
                    return 'M' + sourceX + ',' + sourceY + 'L' + targetX + ',' + targetY;
                }
                else{ // nice cubic bezier curve for the loop ( link with same source and target)
                    return "M" + (d.source.x-2)  +',' + (d.source.y-18) + " c -50,-50 50,-50 10,0  ";
                }
            });
            // draw nodes from proper position
            this.circle.attr('transform', function(d) {
                return 'translate(' + d.x + ',' + d.y + ')';
            });
            // draw label horizontally on links with proper position 
            this.svg.selectAll("text.linklabelholder").attr('transform', function(d) {
                if(d.target != d.source) {
                    return 'translate(' + (d.target.x - (d.target.x - d.source.x)/2) + ',' + (d.target.y - (d.target.y - d.source.y)/2) + ')';
                }
                else { // label on loops
                    return 'translate(' + (d.source.x -5) + ',' + (d.target.y - 60) + ')';
                }
            });
        },
        
        // MOUSE FUNCTIONS
        resetMouseVars: function(){
            this.mousedown_node = null;
            this.mouseup_node = null;
            this.mouseover_node = null;
            this.mousedown_link = null;
        },
        mousedown: function(doc, data, id){
            d3.event.preventDefault();
            this.svg.classed('active', true);
            if(d3.event.ctrlKey || this.mousedown_node || this.mousedown_link) return;
            // insert new node at point
            var point = d3.mouse(doc);
            var node = {id: ++this.lastNodeId, initial: false, final: false, hidden: false};
            node.x = point[0];
            node.y = point[1];
            this.nodes.push(node); // Dont have to set the model because push method hook that
            this.model.set("lastNodeId", this.lastNodeId);
            this.touch();
            
            this.update();
        },
        mousemove: function(doc, data, id){
            d3.event.preventDefault();
            if(!this.mousedown_node) return;
            // update drag line
            this.drag_line.attr('d', 'M' + this.mousedown_node.x + ',' + this.mousedown_node.y + 'L' + d3.mouse(doc)[0] + ',' + d3.mouse(doc)[1]);
            this.update();  
        },
        mouseup: function(doc, data, id){
            
            // add new link when we are dragging from a node
            if(this.mousedown_node) {
                this.drag_line.classed('hidden', true)
                    .style('marker-end','');
                this.mouseup_node = this.mouseover_node;
                if(this.mouseup_node === this.mousedown_node) { this.resetMouseVars(); return; }
                
                // add link and update if already exist
                var source, target;
                    source = this.mousedown_node;
                    target = this.mouseup_node;
                var link;
                link = this.links.filter(function(l) {
                    return (l.source === source && l.target === target);
                })[0];

                if(this.mousedown_node.hidden == false) // a transparent node can't have other links
                {
                    if(!link) {
                        link = {source: source.id, target: target.id, label: 'b', short: false};
                        this.links.push(link);
                        this.model.set('links', this.links)
                        this.touch();
                    }
                }
                
                // Reset variable and select the new link
                this.selected_link = link;
                this.selected_node = null;
                this.mousedown_node = null;
                this.mouseup_node = null;
                this.mousedown_link = null;
                this.update();
            }
            
            // because :active only works in WebKit
            this.svg.classed('active', false);

            // clear mouse event vars
            this.resetMouseVars();
        },
        
        // UPDATE FUNCTION
        update: function(){
            
            var colors = d3.scale.category10();
            
            // closure
            var that = this;

            // Update the datas on model
            this.nodes = this.model.get('nodes');
            this.links = this.model.get('links');
            
            // lastNodeId is not a parameter, he is udpate automatically. But you can still access to it
            this.lastNodeId = this.nodes[this.nodes.length - 1].id;
            this.model.set("lastNodeId", this.lastNodeId);
            this.model.set('nodes', this.nodes);
            this.touch();
            
            // frame with editable text for labels 
            this.frame = this.svg.append('foreignObject')
                                .attr('x', 0)
                                .attr('y', 0)
                                .attr("width", 800)
                                .attr("height", 50)
                                .append("xhtml:form")
                                .attr('class', 'frame')
                                .attr("onSubmit", function(){return false;})
                                .append('input')
                                .on("mouseover", function() {
                                     this.focus();
                                     that.overtext = true;
                                })
                                .attr("value", function() {
                                    if(that.selected_link) { return that.selected_link.label;}
                                })
                                .on("mouseout", function() {
                                    this.blur()
                                    that.overtext = false;
                                    if(that.selected_link) {
                                        var txt = that.frame.node().value;
                                        that.selected_link.label = txt;
                                    }
                                });
            
            // Define elements groups and bound them data
            this.path = this.path.data(this.links);
            this.circle = this.circle.data(this.nodes, function(d) {return d.id;});    
            
            // LINKS
            // Update existing link
            this.path.selectAll('path')
                .classed('selected', function(d) {return d == that.selected_link;})
                .style('marker-end', function(d) {return 'url(#end-arrow)'; })
                .style('stroke-dasharray', function(d) {
                    if (d == that.selected_link) {return "10,2";}
                    else {return'';}})
            // Enter new link
            var p = this.path.enter().append('svg:g');
            p.append('svg:path')
                .attr('class', 'link')
                .style('fill', 'none')
                .style('stroke', '#000')
                .style('stroke-width', '2px')
                .style('cursor', 'pointer')
                .classed('selected', function(d) { return d === that.selected_link; })
                .style('marker-end', function(d) {return 'url(#end-arrow)'; })
                .style('stroke-dasharray', function(d) {
                    if (d == that.selected_link) {return "10,2";}
                    else {return'';}})
                .on('mousedown', function(d) {
                      if(d3.event.ctrlKey) return;
                      // select link
                      that.mousedown_link = d;
                      if(that.mousedown_link === that.selected_link) that.selected_link = null;
                      else that.selected_link = that.mousedown_link;
                      that.selected_node = null;
                      that.update();
                })
            
            // Exit old links
            this.path.exit().remove();
            
            // LINKS LABELS
            // Update existing labels
            this.path.selectAll('text')
                .text(function(d){return d.label});
            // Show link labels
            p.append('svg:text')
            .attr("class", "linklabelholder")
            .style("font-size", "20px")
            .style("fill", "#00CCFF")
            .text(function(d) {return d.label});
            
            // NODES
            // Update existing nodes
            this.circle.selectAll('circle')
                .style('fill', function(d) {
                    if (d.hidden == true) {return "transparent";}
                    else {return (d === that.selected_node) ? d3.rgb(colors(d.id)).brighter().toString() : colors(d.id); }})
            
            // Enter new nodes
            var g = this.circle.enter().append('svg:g')
            g.append('svg:circle')
                .attr('class', 'node')
                .attr('r', 20)
                .style('stroke-width', '1.5px')
                .style('cursor', 'pointer')
                .style('fill', function(d) {
                    if (d.hidden == true) {return "transparent";}
                    else {return (d === that.selected_node) ? d3.rgb(colors(d.id)).brighter().toString() : colors(d.id); }})
                .style('stroke', function(d) {
                    if (d.hidden == true) {return "transparent";}
                    else {return d3.rgb(colors(d.id)).darker().toString(); }})
                .on('mouseover', function(d) {
                    that.mouseover_node = d;
                    d3.select(this).attr('transform', function(d) {
                                        if(d.hidden == true) {return '';}
                                        else{return 'scale(1.5)';}})
                                   .style('stroke', function(d) {
                                        if(d.hidden == true) {return "black";}
                                        else{return d3.rgb(colors(d.id)).darker().toString();}});
                })
                .on('mouseout', function() {
                    console.log('mouseout')
                    d3.select(this).attr('transform', '')
                        .style('stroke', function(d) {
                            if(d.hidden == true) {return "transparent";}
                            else{return d3.rgb(colors(d.id)).darker().toString(); }});
                })
                .on('mousedown', function(d) {
                    d3.event.preventDefault();
                    if(d3.event.ctrlKey) return;
                    // select node
                    that.mousedown_node = d;
                    if(that.mousedown_node === that.selected_node) that.selected_node = null;
                    else {that.selected_node = that.mousedown_node;}
                    that.selected_link = null;
                    // reposition drag line
                    if(that.mousedown_node.hidden == false) // a transparent node cannot have other link
                    {
                        that.drag_line.style('marker-end', 'url(#end-arrow)')
                            .classed('hidden', false)
                            .attr('d', 'M' + that.mousedown_node.x + ',' + that.mousedown_node.y + 'L' + that.mousedown_node.x + ',' + that.mousedown_node.y);
                    }
                    that.update();
                })
               
            // Exit old nodes
            this.circle.exit().remove();
            
            // NODES ID
            // Update nodes ID
            this.circle.selectAll('text')
                .text(function(d) {return d.id;});
            // Show nodes ID
            g.append('svg:text')
                .attr('x', 0)
                .attr('y', 4)
                .attr('class', 'id')
            .style('font', '12px sans-serif')
                .style("text-anchor", "middle")
                .style("font-weight", "bold")
                .style("fill", function(d) {
                    if(d.hidden == true) return "transparent";
                    else return "white";})
                .text(function(d) { return d.id; });
            
            this.force.start();
        }
        // END UPDATE FUNCTION
       
    });
    WidgetManager.register_widget_view("AutomatonView", AutomatonView);
});
```
