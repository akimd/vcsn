# _automaton_.minimize(string algo = "weighted")

Minimizes a automaton

The algorithm can be: 
- Weighted(by default) works with wighted automaton
- Signature, Moore and Brzozowski just work on boolean automaton. With the algorithm Moore: the automaton must have deterministic

Preconditions:
- its labelset is free

Postconditions:
- the result is equivalent to the input automaton
- the result is deterministic

See also:
- [automaton.is_deterministic](automaton.is_deterministic.ipynb)
- [automaton.cominimize](automaton.cominimize.ipynb)

### Examples

```python
import vcsn
a1 = vcsn.automaton('''
  vcsn_context = "lal_char(abc)_z"
  $ -> 0
  0 -> 1 <2>a
  0 -> 2 <3>a
  1 -> 1 a
  1 -> 3 <4>a
  2 -> 2 a
  2 -> 4 a
  3 -> $
  4 -> $
''', 'daut')
a1
```

### Weighted

Using minimize or minimize("weightd")

```python
a1.minimize()
```

```python
a2 = vcsn.automaton('''
  vcsn_context = "lal_char(abcde)_b"
  $ -> 0
  0 -> 1 a
  0 -> 3 b
  1 -> 1 a
  1 -> 2 b
  2 -> 2 a
  2 -> 5 b
  3 -> 3 a
  3 -> 4 b
  4 -> 4 a
  4 -> 5 b
  5 -> 5 a
  5 -> 5 b
  5 -> $
''', 'daut')
a2
```

### Signature

```python
a2.minimize("signature")
```

### Moore

```python
a2.is_deterministic()
```

```python
a2.minimize("moore")
```

### Brzozowski

```python
a2.minimize("brzozowski")
```
